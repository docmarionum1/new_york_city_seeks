```python
def write_haiku(words):
    words = ast.literal_eval(words)
    haiku = [[],[],[]]
    i = 0

    for word in words:
        if word == 'i++':
            i += 1
        elif word == 'i--':
            i -= 1
        elif word[0] == '-':
            while (len(haiku[i][-1]) > 0):
                haiku[i][-1] = haiku[i][-1][:-1]
            haiku[i].pop()
        else:
            haiku[i].append('')
            for c in word:
                haiku[i][-1] += c

    return '\n'.join([' '.join(line) for line in haiku])
```

```python
def create_new_path(row):
    path = ast.literal_eval(row['path'])
    haiku = row['haiku2'].split(' ')
    #print(path,haiku)
    loc = 0
    path2 = []
    for i,word in enumerate(path):
        #print(i,word,loc)
        if word[0] == '-':
            path2.append('-')
            loc -= 1
        elif word in ['i++', 'i--']:
            path2.append(word)
        elif loc < len(haiku) and word.lower() in haiku[loc].lower():
            path2.append(haiku[loc])
            loc += 1
        else:
            path2.append(word)
            loc += 1
        
    return str(path2)#, write_haiku(path2)

create_new_path(df.iloc[4])
```

```python
df['path2'] = df.apply(create_new_path, axis=1)
df
```

```python
df['haiku3'] = df['path2'].apply(write_haiku)
df
```

```python
df[['title', 'haiku3', 'path2']].rename(columns={'path2': 'path', 'haiku3': 'haiku'}).to_csv('results_filtered/edited.csv', index=False)
```
