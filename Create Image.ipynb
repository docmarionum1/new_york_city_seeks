```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

#astr = '''The rain in Spain falls mainly on the plains.'''
#para = textwrap.wrap(astr, width=15)

#haiku = ["Monitoring the", "workload and evaluate", "work performances"]
haiku = ["COMPUTER SYSTEMS MANAGER", "The new york city", "transplants trees shrubs and plants as", "is not required"]

MAX_W, MAX_H = 696, 200
im = Image.new('RGB', (MAX_W, MAX_H), (255, 255, 255, 0))
draw = ImageDraw.Draw(im)
#font = ImageFont.truetype(
#    '/usr/share/fonts/truetype/msttcorefonts/Arial.ttf', 18)
font = ImageFont.truetype("arial.ttf", 30)

#current_h, pad = 10, 10
pad = 10
title_pad = 20
line_pad = 50
total_h = sum([draw.textsize(line, font=font)[1] + pad for line in haiku]) + (title_pad - pad)
current_h = (MAX_H - total_h)/2

title = haiku[0]
w,h = draw.textsize(title, font=font)
draw.text(((MAX_W - w) / 2, current_h), title, font=font, fill=(0,0,0))
current_h += h + title_pad

draw.line([(line_pad, current_h-title_pad/2), (MAX_W - line_pad, current_h-title_pad/2)], fill=(0,0,0))

for line in haiku[1:]:
    w, h = draw.textsize(line, font=font)
    draw.text(((MAX_W - w) / 2, current_h), line, font=font, fill=(0,0,0))
    current_h += h + pad
    

im
```
