```python
import pandas as pd
```

```python
df = pd.read_csv('results_filtered/unedited.csv')
df['haiku2'] = df['haiku'].str.replace(r'\s', ' ').str.lower()
df['haiku2'] = df['haiku2'].str.replace('  ', ' ')

df['haiku2'] = df['haiku2'].str.replace('i-t', 'IT')
for acronym in [
    'hvac', 'dep ', ' pm ', ' tb '
]:
    df['haiku2'] = df['haiku2'].str.replace(acronym, acronym.upper())

for proper_noun in [
    ' i ',
    'outlook', 'excel ', 'word', 'power point', 'microsoft', 'access', 'powerpoint',
    'oracle', 'unix', 'linux', 'solaris', 'windows', 'perl',
    'new york city', 'new york state', 'new yorkers', 'new york', 'bureau of tuberculosis', 
    'department of youth and communities', 'bureau of water and sewer projects',
    'bureau of mental health', 'east river', 'ddc',
    'department of health'
]:
    print(proper_noun, proper_noun.title())
    df['haiku2'] = df['haiku2'].str.replace(proper_noun, proper_noun.title())
    
#df['haiku2'] = df['haiku2'].str.capitalize() + '.'
df['haiku2'] = df['haiku2'].str.slice(0,1).str.upper() + df['haiku2'].str.slice(1) + '.'

df
```

```python
df[['title', 'haiku', 'haiku2', 'path']].to_csv('results_filtered/partially_edited.csv', index=False)
```
