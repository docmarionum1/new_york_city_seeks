```python
import json
import random
import re
import pathlib
import pandas as pd
```

```python
dictionary = {}
with open('phoneme-groups-with-syllables.json') as f:
    for l in f:
        j = json.loads(l)
        if len(j['word']) > 1 or j['word'] in ['A', 'I']:
            dictionary[j['word']] = len(j['syllables'])
```

```python
jobs_df = pd.read_csv('jobs.csv')
jobs_df
```

```python
special_chars = re.compile(r'[,()/$\'"*]|(- )')
whitespace = re.compile(r'\s+')
sentence_enders = re.compile(r'[.?!;•:]')
def clean_string(s):
    s = s.replace('IT', 'I-T').replace('I.T.', 'I-T')
    a = special_chars.sub(' ', s)
    b = whitespace.sub(' ', a)
    c= sentence_enders.sub('.', b)

    return c

s = jobs_df.iloc[1]['Job Description']
#s = jobs_df.iloc[11]['Preferred Skills']
print(s)
print()
print(clean_string(s))
```

```python
words = {}
for i,row in jobs_df.iterrows():
    for col in ['Job Description', 'Preferred Skills']:
        try:
            s = row[col].encode('windows-1252').decode('utf-8')
        except:
            s = row[col]
        s = clean_string(s)
        for sentence in s.split('.'):
            for word in sentence.strip().split(' '):
                u = word.upper()
                if u not in words:
                    words[u] = 1
                else:
                    words[u] += 1
```

```python
words.items()
```

```python
words_df = pd.DataFrame(list(words.items()), columns=['word', 'count'])
words_df['in_dict'] = words_df['word'].apply(lambda x: x in dictionary)
words_df
```

```python
for i,row in words_df[~words_df['in_dict'] & (words_df['count'] > 25)].sort_values('count', ascending=False).iterrows():
    print(row['word'])
```

```python
words_df[~words_df['in_dict'] & (words_df['count'] > 10)].sort_values('count', ascending=False)[['count', 'word']].to_csv('out_of_dict_words.csv', index=False)
```
